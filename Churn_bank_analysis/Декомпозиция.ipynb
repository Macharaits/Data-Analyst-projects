# Декомпозиция

## Банки — Анализ оттока клиентов

**Описание проекта:** 

В распоряжении данные о клиентах банка. Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск. Клиенты стали уходить, необходимо провести анализ и разобраться в причинах оттока.

## Шаг 1. Загрузка и предоработка данных

- Загрузка датафрейма. Проверка корректность названий колонок и корректировка при необходимости;
- Проверить соответствие типов и изменить при необходимости;
- Изучение пропущенных значений. Заполнение и удаление пропусков;
- Исследовать дубликаты. Удалить дубликаты;

## Шаг 2. Исследовательский анализ данных (EDA)

- Рассмотрим колонки по отдельности, при необходимости построим графики;
- Средние значения признаков в двух группах — тех, кто ушел в отток и тех, кто остался
- Построим столбчатые гистограммы и распределения признаков для тех, кто ушёл (отток) и тех, кто остался (не попали в отток)
- Построим матрицу корреляций и отобразим её.

## Шаг 3. Выявление портреты клиентов, которые склонных уходить из банка

- Стандартизируем данные
- Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов

## Шаг 4. Формулировка и проверка с

- Проверка гипотезы различия дохода между теми клиентами, которые ушли и теми, которые остались.
- Проверка гипотезы различия количества продуктов  между теми клиентами, которые ушли и теми, которые остались.
